```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df = df[(df['year']>=2010) & (df['transmission']!='Other') & (df['fuelType']!='Other')]
df
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.displot(data=df, x='price', col='transmission', height=5, aspect=2, )
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', height=5, aspect=2, multiple='stack')
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, kind='kde', x='price' ,height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price' , kde=True, height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='transmission', fill=True)
plt.show()
```

```python
sns.displot(data=df, x='price', y='year' , kind='kde', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price' , kind='ecdf', height=5, aspect=2)
plt.grid(which='both') #'major', 'minor', 'both' qiymatlar berish mumkin
plt.minorticks_on() #setkani kichraytirib beradi (ma'lumotlarni aniqroq tahlil qilish uchun qo'l keladi)
plt.show()
```

```python
sns.displot(data=df, x='price' , kind='ecdf', height=5, aspect=2)
plt.grid(which='major') #'major', 'minor', 'both' qiymatlar berish mumkin
plt.minorticks_on() #setkani kichraytirib beradi (ma'lumotlarni aniqroq tahlil qilish uchun qo'l keladi)
plt.show()
```
