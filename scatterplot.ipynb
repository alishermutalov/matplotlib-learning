```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df.head()
```

```python
df = df[df.year>1999]
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.regplot(data=df, x='year', y='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.regplot(data=df, x='mileage', y='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='mileage', y='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', style='fuelType', s=100)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```
